# Creating Loyalty Flag

```python
# transform the ‘order_number’ column by applying the max() function from the NumPy library to create new column

ords_prods_merge['max_order'] = ords_prods_merge.groupby('user_id')['order_number'].transform('max')
```

```python
# check output

ords_prods_merge [['user_id', 'order_number', 'max_order']].head(60)
```

```python
# create a loyalty flag base on order numbers using loc function with if-statement

ords_prods_merge.loc [ords_prods_merge['max_order'] >= 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc [ (ords_prods_merge['max_order'] < 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# check flag output

ords_prods_merge [['user_id' , 'max_order' , 'loyalty_flag']].head (60)
```

```python
# check frequency of the new derived column

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# basic statistics of each categories in loyalty flag

ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'median', 'min' , 'max'] })
```

# Creating Spending Flag

```python
# transform the ‘price’ column by applying the mean() function from the NumPy library to create new column

ords_prods_merge['avg_order_price'] = ords_prods_merge.groupby('user_id') ['prices'].transform('mean')
```

```python
# check output

ords_prods_merge[['user_id','loyalty_flag','avg_order_price']]
```

```python
# create spending flag

ords_prods_merge.loc[ords_prods_merge['avg_order_price'] < 10 , 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.loc [ords_prods_merge['avg_order_price'] >=10 , 'spending_flag'] = 'High spender'
```

```python
# check frequency

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# check output

ords_prods_merge[['user_id' , 'spending_flag' , 'loyalty_flag']]
```

# Create Frequency Flag

```python
# transform the ‘days_since_last_order’ column by applying the median() function from the NumPy library to create new column

ords_prods_merge['median_of_regularity'] = ords_prods_merge.groupby('user_id') ['days_since_prior_order'].transform('median')
```

```python
# check output

ords_prods_merge[['user_id', 'days_since_prior_order', 'median_of_regularity']].head(20)
```

```python
# create frequency flag

ords_prods_merge.loc[ords_prods_merge['median_of_regularity'] > 20 , 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_of_regularity'] <= 20) & (ords_prods_merge['median_of_regularity'] >10) , 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ ords_prods_merge['median_of_regularity'] <= 10 , 'order_frequency_flag'] = 'Frequent customer'
```

```python
# check flag frequency

ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# check the null values in 'order_frequency_flag' column

ords_prods_merge.loc[ords_prods_merge ['order_frequency_flag'].isnull()]
```
