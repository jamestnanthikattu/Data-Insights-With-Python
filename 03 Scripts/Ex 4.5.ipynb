# Observation: Explanatory Analysis (Products Dataframe)
An explanatory analysis of the df_products dataframe revealed the presence of outliers in the data. The prices column has a mean of 9.99 and a standard deviation of 453.6, which is more than two standard deviations away from the mean, indicating outlier values. By identifying all entries in the prices column that are more than two standard deviations from the mean, two outliers were found: prices of 14900 and 99999.

These outliers were replaced with the median value of 7.1. Given the wide range of values in the column, with a minimum of 1 and a maximum of 99999, it is advisable to replace outliers with the median rather than the mean. The median is less affected by extreme values, making it a more suitable measure for handling such outliers. After replacing the outliers, the mean of the prices column decreased to 7.6.
